Hi, in this article, I want to find out 'who is the longest buried writer in history'?

The method I used is using the artworks' acquisition Year to minus the artists' pass away Year.

```python
plaimport pandas as pd
```

```python
pd.options.display.max_rows = 100
```

```python
artist = pd.read_csv('C:/Users/pc/Desktop/artist_data.csv', delimiter=",", encoding='utf-8').rename(columns={'name':'artist'})
artist
```

HERE I renamed the column 'name' to 'artist' in order to find a same join name with the other csv.

```python
artwork = pd.read_csv('C:/Users/pc/Desktop/artwork_data.csv', low_memory=False, delimiter=',',encoding='utf-8')
artwork
```

Then merge them together.

```python
merged = pd.merge(artist, artwork, on='artist')
merged
```

We got the merged dataset, so we may visualize the trend first.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
sns.relplot(data=merged, x='yearOfDeath', y='acquisitionYear')
plt.show()
```

With the development of technology, more art pieces were found around 2000.

A very large proportion of the artists' paintings became famous after his death (if acquisitionYear equals the year of fame)

We can calculate that.

```python
timelength = merged['acquisitionYear'] - merged['yearOfDeath']
timelength = pd.DataFrame(timelength)
timelength.columns=['timegap']
timelength
```

```python
df1 = pd.concat([merged['artist'], timelength['timegap']],axis=1, keys=['artist','timegap'])
df1
```

```python
gkk = df1.groupby(['artist'])['timegap'].sum().reset_index().rename(columns={'timegap':'total_year'})
gkk
```

```python
ggg = df1.groupby(['artist'])['timegap'].count().reset_index()
ggg = ggg.rename(columns={'timegap':'piece_num'})
ggg
```

```python
merged2 = pd.merge(gkk, ggg, on='artist')
merged2['ave_gap'] = merged2['total_year'] / merged2['piece_num']
merged2['rating_rank'] = merged2['ave_gap'].rank(ascending=0)
merged2 = merged2.set_index('rating_rank')
merged2 = merged2.sort_index()
merged2.head(100)
```

```python
print(merged2.dtypes)
```

```python
after_death = merged2['ave_gap'][merged2['ave_gap'] >= 0].count()
print(after_death)
```

```python
before_death = merged2['ave_gap'][merged2['ave_gap'] < 0].count()
print(before_death)
```

From this calculation, it is not difficult to see that 1248 artists became famous, while 907 artists had already become famous before they passed away.
